# Image Processing

Image processing is a field of computer science and engineering that focuses on performing operations on images to:

* Enhance them (make them clearer or more visually appealing),

* Analyze them (extract meaningful information), or

* Transform them (convert from one form to another).

It plays a crucial role in applications like medical imaging, facial recognition, machine vision, remote sensing, and AI-based perception systems.

| Task                               | Description                                                                       |
| ---------------------------------- | --------------------------------------------------------------------------------- |
| **Filtering**                      | Enhance or suppress features (e.g., edge detection, blurring, sharpening).        |
| **Resizing**                       | Scale images up or down (used in web and machine learning pipelines).             |
| **Denoising**                      | Remove unwanted noise (grain or distortion).                                      |
| **Thresholding**                   | Convert images to black & white based on intensity values (used in segmentation). |
| **Edge Detection**                 | Identify edges or boundaries in images (e.g., using Sobel, Canny filters).        |
| **Color Space Conversion**         | Change between RGB, grayscale, HSV, etc.                                          |
| **Morphological Operations**       | Modify the structure of objects in binary images (e.g., dilation, erosion).       |
| **Object Detection / Recognition** | Locate or identify objects (e.g., face detection, number recognition).            |


# ü§ñ What is a Convolutional Layer?
A Convolutional Layer (often written as Conv2D) is a core building block of a Convolutional Neural Network (CNN). It is designed to automatically and adaptively learn spatial features (like edges, textures, patterns) from images.

A convolutional layer takes an image and applies a set of small filters (kernels) to detect features like edges, curves, or textures ‚Äî then passes this info to deeper layers to build more complex understanding (like eyes, digits, faces, etc).

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create a 10x10 image representing the number '9'
image = np.array([
    [0,   0,   0,   0,   255, 255, 255, 255, 0,   0],
    [0,   0,   0,   255, 0,   0,   0,   255, 0,   0],
    [0,   0,   255, 0,   0,   0,   0,   255, 0,   0],
    [0,   0,   0,   255, 0,   0,   0,   255, 0,   0],
    [0,   0,   0,   0,   255, 0,   0,   255, 0,   0],
    [0,   0,   0,   0,   0,   255, 255,   255, 0,   0],
    [0,   0,   0,   0,   0,   0,   0, 255, 0,   0],
    [0,   0,   0,   0,   0,   0,   0,   255, 0,   0],
    [0,   0,   0,   0,   0,   0,   0,   255, 0,   0],
    [0,   0,   0,   0,   0,   0,   0,   255, 0,   0]
])

# Plot the image
plt.imshow(image, cmap='gray', interpolation='nearest')
plt.title('10x10 Image of the Number 9')
plt.axis('off')  # Hide the axis
plt.show()
image
```

```python
# Step 2: Normalize the image by dividing by 255
normalized_image = image / 255.0

# Plot the normalized image
plt.imshow(normalized_image, cmap='gray', interpolation='nearest')
plt.title('Normalized 10x10 Image of the Number 9')
plt.axis('off')  # Hide the axis
plt.show()
normalized_image
```

# Conv2D Layer

Conv2D stands for 2-dimensional convolution ‚Äî it's one of the most fundamental operations in Convolutional Neural Networks (CNNs) used in image processing and deep learning. It‚Äôs how neural networks ‚Äúsee‚Äù patterns like edges, textures, corners, and even complex shapes in images.

ai24.png

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import convolve2d

# Step 3: Define a kernel (Laplacian-like)
loop_kernel = np.array([
    [0,  1, 0],
    [1, -4, 1],
    [0,  1, 0]
])

# Apply convolution using 2D convolution
convolved_output = convolve2d(normalized_image, loop_kernel, mode='same', boundary='fill', fillvalue=0)

# Plot normalized image
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.imshow(normalized_image, cmap='gray', interpolation='nearest')
plt.title('Step 2: Normalized Image')
plt.axis('off')

# Plot convolved output
plt.subplot(1, 2, 2)
plt.imshow(convolved_output, cmap='hot', interpolation='nearest')
plt.title('Step 3: After Loop Detection Kernel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
convolved_output
```

# Maxpooling

**Max pooling** is a downsampling operation used in convolutional neural networks (CNNs) to reduce the spatial dimensions (width and height) of feature maps. It works by sliding a window (e.g., 2√ó2) over the input and selecting the **maximum value** in each region. This helps retain the most important features (like edges or textures) while reducing computation, overfitting, and model size. Max pooling introduces **translation invariance**, meaning small shifts in the input won‚Äôt drastically change the output. It is commonly used after convolution layers to compress the data and highlight dominant features for further processing in deep learning models.

ai25.jfif

```python
from skimage.measure import block_reduce

# Apply 2x2 max pooling
pooled_output = block_reduce(convolved_output, block_size=(2, 2), func=np.max)

# Show original convolved output and max pooled result
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Original convolved image
plt.subplot(1, 2, 1)
plt.imshow(convolved_output, cmap='hot', interpolation='nearest')
plt.title('Convolved Output (Before Pooling)')
plt.axis('off')

# Max pooled image
plt.subplot(1, 2, 2)
plt.imshow(pooled_output, cmap='hot', interpolation='nearest')
plt.title('Max Pooled Output (2x2)')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
pooled_output
```

```python
from scipy.signal import convolve2d
second_kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])


# Apply second 3x3 convolution on the pooled output
second_convolved_output = convolve2d(pooled_output, second_kernel, mode='same', boundary='fill', fillvalue=0)

# Plot the second convolution result
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Pooled image (input to second conv)
plt.subplot(1, 2, 1)
plt.imshow(pooled_output, cmap='hot', interpolation='nearest')
plt.title('Input to 2nd Conv (Max Pooled)')
plt.axis('off')

# Output of second convolution
plt.subplot(1, 2, 2)
plt.imshow(second_convolved_output, cmap='hot', interpolation='nearest')
plt.title('After 2nd Conv2D')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
second_convolved_output
```

```python
from skimage.measure import block_reduce
import matplotlib.pyplot as plt
import numpy as np

# Perform 2x2 max pooling on the combined edge map
pooled_edges = block_reduce(second_convolved_output, block_size=(2, 2), func=np.max)

# Plot original edge map vs pooled version
plt.figure(figsize=(10, 4))

# Original edge map (from Sobel)
plt.subplot(1, 2, 1)
plt.imshow(second_convolved_output, cmap='hot')
plt.title('Combined Edge Map (Before Pooling)')
plt.axis('off')

# After 2x2 Max Pooling
plt.subplot(1, 2, 2)
plt.imshow(pooled_edges, cmap='hot')
plt.title('After 2√ó2 Max Pooling')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
pooled_edges
```

```python
# Flatten the pooled edge output to a 1D vector
flattened_output = pooled_edges.flatten()

# Plot the flattened vector as a bar chart
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
plt.bar(range(len(flattened_output)), flattened_output, color='orange')
plt.title('Flattened Output Vector (After Max Pooling)')
plt.xlabel('Index')
plt.ylabel('Activation Value')
plt.grid(True)
plt.show()
```

```python
flattened_output
```
