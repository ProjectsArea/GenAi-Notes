#Metrics Definition:
In machine learning (ML), metrics are numbers that help us understand how well a model is working. They measure how accurate or correct the model's predictions are. For example, if a model predicts whether an email is spam or not, a metric like accuracy tells us how many times it got the prediction right. Other metrics like precision, recall, and F1-score give more details about performance, especially when the data is imbalanced. These metrics help us compare models, find mistakes, and choose the best one for the task. Simply put, metrics show how good (or bad) a machine learning model is.

#a.Confusion Matrix:

A **confusion matrix** is a table used to evaluate the performance of a **classification model** by comparing actual and predicted labels. It shows how many predictions were correct and where the model made errors. The matrix has four main parts: **True Positives (TP)**, **True Negatives (TN)**, **False Positives (FP)**, and **False Negatives (FN)**. For example, in binary classification, TP means the model correctly predicted the positive class. It helps calculate metrics like **accuracy**, **precision**, **recall**, and **F1-score**. A confusion matrix provides a clear, detailed view of model performance beyond a single number like accuracy.

#**d.Recall**
In machine learning, recall is a performance metric used to measure how well a model can identify all relevant instances, especially in classification problems. It is the ratio of correctly predicted positive observations to all the actual positives. In other words, recall tells you how many of the actual positive cases the model was able to find.

#**e.F1 Score**
The **F1 score** is a way to measure how well a machine learning model is performing by combining **precision** and **recall**. Precision shows how many of the predicted positive results were actually correct, while recall tells us how many actual positive cases the model correctly found. The F1 score takes both into account and gives a single number that balances these two metrics. Itâs especially helpful when thereâs an imbalance in the data, meaning one type of result (like positive cases) is rarer than the other. A higher F1 score means the model is doing well in both areas.

#**Support Vector Machine**
Support Vector Machine (SVM) is a machine learning algorithm used to classify data into different categories. It works by drawing a line (or boundary) that separates the data points of different categories in the best possible way. The goal is to make sure the distance between the line and the closest data points (called support vectors) is as large as possible. This helps the model make better predictions. SVM can handle both simple and complex data patterns and is used in tasks like image recognition, spam detection, and medical diagnosis because it's accurate and reliable.

#**KNN (K Nearest Neighbours)**
The **K-Nearest Neighbors (KNN)** algorithm is a simple, supervised machine learning method used for classification and regression. It works by comparing a new data point to existing data points and finding the **'k' closest neighbors** based on distance (like Euclidean distance). The new point is then classified based on the most common category among its neighbors (for classification) or averaged values (for regression). KNN doesnât learn from the data during trainingâinstead, it stores the entire dataset and makes decisions at prediction time. Itâs easy to understand and use, but can be slow with large datasets or high dimensions.

#**Decision Trees**
A **Decision Tree** is a supervised machine learning algorithm used for both classification and regression tasks. It works like a flowchart, where each internal node represents a decision based on a feature, each branch shows the outcome of the decision, and each leaf node represents the final prediction or class label. The tree splits the data into smaller and smaller groups based on questions about the features, aiming to make the groups as pure as possible. Decision trees are easy to understand and visualize, but they can overfit if the tree is too deep. Techniques like pruning help prevent this.
# Structure of decision tree


Root Node: The starting point that represents the entire dataset.

Decision/Internal Nodes: Represent tests or decisions based on features.

Branches: Outcomes of the decisions.

Leaf Nodes: Final output labels (classes or values).
